```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig
```

```python
n = np.linspace(-5, 10, 150)

f = sig.unit_impulse(10, [0, 2, 4])
g = 4 * (np.heaviside(-n, 1) - np.heaviside(-n + 5, 1))
y = np.convolve(g, f)

plt.figure(figsize=(12, 8))
plt.subplot(211)
plt.plot(f)
plt.plot(n, g)
plt.title('F(t) and G(t)')

plt.figure(figsize=(10, 8))
plt.subplot(212)
plt.plot(y)
plt.title('Convolution of F(t) and G(t)')

plt.tight_layout()
```
