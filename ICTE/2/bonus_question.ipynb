### Periodic signals

1. $x[ùëõ] = \cos (1.2\pi ùëõ)$
2. $ùë¶(ùë°) = \cos(2ùë° ‚àí \pi)$
3. $ùë¶[n] = \cos(2n ‚àí \pi)$

### Aperiodic signals

1. $ùë• ( ùë° ) = \cos (1.2\pi ùë°)ùë¢(ùë°)$
2. $z(t) = e ^{-t} \cos (2t)$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np

n = np.linspace(-5, 5, 100)
```

$x[ùëõ] = \cos (1.2\pi ùëõ)$

```python
x1 = np.cos(1.2 * np.pi * n)

plt.figure(figsize=(10, 8))
plt.stem(n, x1)
```

$ùë¶(ùë°) = \cos(2ùë° ‚àí \pi)$

```python
y1 = np.cos(2 * n - np.pi)

plt.figure(figsize=(10, 8))
plt.plot(n, y1)
```

$ùë¶[n] = \cos(2n ‚àí \pi)$

```python
y2 = np.cos(2 * n - np.pi)

plt.figure(figsize=(10, 8))
plt.stem(n, y2)
```

$ùëß[ùëõ] = (‚àí1) ^n$

```python
n1 = np.linspace(-10, 10, 1000)
z1 = np.power(-1, n + 0j)

plt.figure(figsize=(12, 8))
plt.stem(n, z1)
```

$ùë• ( ùë° ) = \cos (1.2\pi ùë°)ùë¢(ùë°)$

```python
u = np.heaviside(n, 1)
x2 = np.cos(1.2 * np.pi * n) * u

plt.figure(figsize=(10, 8))
plt.plot(n, x2)
```

$z(t) = e ^{-t} \cos (2t)$

```python
z2 = np.e ** (-n) * np.cos(2 * n)

plt.figure(figsize=(10, 8))
plt.plot(n, z2)
```
